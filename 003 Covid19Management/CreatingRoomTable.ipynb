```python
import pandas as pd

roomDistribution = pd.read_csv("pre__hospitalRoom.csv")
roomDistribution

outerList = []
innerList = []
columnName = list(roomDistribution.columns)

row = 0
while row <= 12:
    
    col = 1
    while col <= 7:
        
        print("I am running")
        
        firstBracket = columnName[col].find("(")
        roomType = (columnName[col])[0:firstBracket]
        roomCapacity = ((columnName[col])[firstBracket+1])

        innerList = []
        
        innerList.insert(0,roomDistribution.iloc[row,0])
        innerList.insert(1,roomType)
        innerList.insert(2,roomCapacity)
        
        iterate = roomDistribution.iloc[row,col]
        
        print("Iterate is : ", str(iterate))
        
        sequence = 0;
        while iterate > 0:
            
            sequence += 1;
            
            id = str(sequence).zfill(3)
            
            if roomType == "HDU":
                id = "H" + str(roomCapacity) + "-" + id
            elif roomType == "ICU":
                id = "I" + str(roomCapacity) + "-" +id
            else:
                id = "N" + str(roomCapacity) + "-" +id
               
            print("-------",id)
            
            tempList = [innerList[0],innerList[1],innerList[2],id]

            outerList.append(tempList)
                                    
            print("I am iterating")
    
            iterate -= 1
        col += 1
    row += 1

finalProduct = pd.DataFrame(outerList)

finalProduct.to_csv("Room.csv")

print("I am done")

finalProduct
```
